```python
# Step 2: Imports
import pandas as pd
import torch
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from google.colab import files
```

```python
# Step 3: Upload the CSV file manually
uploaded = files.upload()  # This will let you select and upload your local file
```

```python
# Step 4: Load your dataset (adjust filename if needed)
df = pd.read_csv("comments_with_scores.csv")
df = df.dropna(subset=['clean_comment'])  # Ensure no nulls
df['clean_comment'] = df['clean_comment'].astype(str)
```

```python
# Step 5: Load 768-D model (all-mpnet-base-v2)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer('all-mpnet-base-v2', device=device)
```

```python
# Step 6: Generate embeddings (batch-wise with progress bar)
embeddings = model.encode(df['clean_comment'].tolist(), batch_size=32, show_progress_bar=True)
```

```python
# Step 7: Convert to DataFrame
embed_df = pd.DataFrame(embeddings)
embed_df['token_ticker'] = df['token_ticker'].values
embed_df['user_name'] = df['user_name'].values
embed_df['comment_id'] = df['comment_id'].values
```

```python
# Step 8: Save to CSV
embed_df.to_csv("comment_embeddings_768D.csv", index=False)
print("🎉 Embeddings saved as comment_embeddings_768D.csv!")
files.download("comment_embeddings_768D.csv")
```
